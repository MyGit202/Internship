# Q.10:-Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

```python
import selenium
import pandas as pd
from selenium import webdriver as driver


driver = driver.Chrome("chromedriver.exe")


url = "https://www.hostelworld.com/"
driver.get(url)
```

```python
hostel_name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
hostel_name


Hostel_name = []
for i in hostel_name:
    Hostel_name.append(i.text)
Hostel_name
```

```python
distance_from_city_centre = driver.find_elements_by_xpath("//span[@class='description']")
distance_from_city_centre


Distance_from_city_centre = []
for i in distance_from_city_centre:
    Distance_from_city_centre.append(i.text)
Distance_from_city_centre
```

```python
ratings = driver.find_elements_by_xpath("//div[@class='score orange big']")
ratings


Ratings = []
for i in ratings:
    Ratings.append(i.text)
Ratings
```

```python
total_reviews = driver.find_elements_by_xpath("//div[@class='reviews']")
total_reviews


Total_reviews = []
for i in total_reviews:
    Total_reviews.append(i.text)
Total_reviews
```

```python
overall_reviews = driver.find_elements_by_xpath("//div[@class='keyword']")
overall_reviews


Overall_reviews = []
for i in overall_reviews:
    Overall_reviews.append(i.text)
Overall_reviews
```

```python
private_for_price = driver.find_elements_by_xpath("//p[@class='no-price body-3 no-privates']")
private_for_price


Private_for_price = []
for i in private_for_price:
    Private_for_price.append(i.text)
Private_for_price
```

```python
dorms_from_price = driver.find_elements_by_xpath("//div[@class='price title-5']")
dorms_from_price


Dorms_from_price = []
for i in dorms_from_price:
    Dorms_from_price.append(i.text)
Dorms_from_price
```

```python
facilities = driver.find_elements_by_xpath("//div[@class='facilities-label facilities']")
facilities


Facilities = []
for i in facilities:
    Facilities.append(i.text)
Facilities
```

```python
# make DataFrame with the help of Pandas

Hostel_details=pd.DataFrame({})
Hostel_details['Hostel_name']= Hostel_name[:20]
Hostel_details['Distance_from_city_centre']= Distance_from_city_centre[:20]
Hostel_details['Ratings']= Ratings[:20]
Hostel_details['Total_reviews']= Total_reviews[:20]
Hostel_details['Overall_reviews']= Overall_reviews[:20]
Hostel_details['Private_for_price']= Private_for_price[:20]
Hostel_details['Dorms_from_price']= Dorms_from_price[:20]
Hostel_details['Facilities']= Facilities[:20]

Hostel_details
```
