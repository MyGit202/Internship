```python
# lets import all the required library
import selenium
import pandas as pd
from selenium import webdriver as driver
import time
from selenium.common.exceptions import NoSuchElementException           # Importing Exception
```

```python
# lets first connect to the webdriver
driver = driver.Chrome("chromedriver.exe")
```

```python
url = "https://www.amazon.in/"
driver.get(url)
```

```python
search_attrs=driver.find_element_by_name("field-keywords")
search_attrs
```

```python
# use this function to write the job title on job page automatically.
search_attrs.send_keys("Laptop")
```

```python
# do click using xpath function
search_btn = driver.find_element_by_xpath('//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]')
search_btn.click()
```

```python
#locating the core i7 filter

filter_button=driver.find_elements_by_xpath("//i[@class='a-icon a-icon-checkbox']")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
```

```python
# Intel Core i9 is not avail in the list of filter therefore i filtered on core i5
filter_button=driver.find_elements_by_xpath("//i[@class='a-icon a-icon-checkbox']")
for i in filter_button:
    if i.text=='Intel Core i5':
        i.click()
        break
```

```python
# Extract all the webelement through xpath.
title = driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")*4
title
```

```python
Title = []
for i in title:
    Title.append(i.text)
Title
```

```python
ratings = driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-beside-button a-text-bold']")*100
ratings
```

```python
Ratings = []
for i in ratings:
    Ratings.append(i.text)
Ratings
```

```python
price = driver.find_elements_by_xpath("//span[@class='a-price-whole']")*5
price
```

```python
Price = []
for i in price:
    Price.append(i.text)
Price
```

```python
# make DataFrame with the help of Pandas

Laptop_Details=pd.DataFrame({})

Laptop_Details['Title']= Title[:100]
Laptop_Details['Ratings']= Ratings[:100]
Laptop_Details['Price']= Price[:100]

Laptop_Details
```
